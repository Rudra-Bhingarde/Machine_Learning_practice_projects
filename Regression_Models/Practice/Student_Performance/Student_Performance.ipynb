# Imports

```python
import numpy as np
np.set_printoptions(threshold= np.inf)
import pandas as pd
```

```python
ds = pd.read_csv('Student_Performance.csv')
```

# Data Preprocessing

```python
X = ds.iloc[:,:-1].values
Y = ds.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
X[:,2]=LE.fit_transform(X[:,2])
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y,test_size=0.25,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
Xtrain[:,[0,1,3,4]]=ss.fit_transform(Xtrain[:,[0,1,3,4]])
Xtest[:,[0,1,3,4]]=ss.transform(Xtest[:,[0,1,3,4]])
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regressorLR = LinearRegression()
```

```python
regressorLR.fit(Xtrain,Ytrain)
```

```python
YpredLR = regressorLR.predict(Xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Ytest,YpredLR)
```

```python
comparison = np.concatenate((YpredLR.reshape(len(YpredLR), 1), Ytest.reshape(len(Ytest), 1)), axis=1)
print(comparison[:50])
```

# Support Vector Regression

```python
from sklearn.svm import SVR
```

```python
regressorSVR = SVR(kernel='rbf')
```

```python
regressorSVR.fit(Xtrain,Ytrain)
```

```python
YpredSVR = regressorSVR.predict(Xtest)
```

```python
r2_score(Ytest,YpredSVR)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressorDTR = DecisionTreeRegressor(random_state=42)
```

```python
regressorDTR.fit(Xtrain,Ytrain)
```

```python
YpredDTR = regressorDTR.predict(Xtest)
```

```python
r2_score(Ytest,YpredDTR)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressorRFR = RandomForestRegressor(n_estimators = 100, random_state=42)
```

```python
regressorRFR.fit(Xtrain,Ytrain)
```

```python
YpredRFR = regressorRFR.predict(Xtest)
```

```python
r2_score(Ytest,YpredRFR)
```
