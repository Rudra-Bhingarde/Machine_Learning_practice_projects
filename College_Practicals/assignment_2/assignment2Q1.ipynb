```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
X = df.iloc[:,0].values
Y = df.iloc[:,-1].values
```

```python
plt.scatter(X,Y,color='blue')
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y,test_size=0.25,random_state = 42)
```

```python
LR = LinearRegression()
```

```python
LR.fit(Xtrain.reshape(-1,1),Ytrain.reshape(-1,1))
```

```python
Ypred = LR.predict(Xtest.reshape(-1,1))
```

```python
mean_squared_error(Ypred,Ytest)
```

```python
mean_absolute_error(Ypred,Ytest)
```

```python
r2_score(Ypred,Ytest)
```

```python
plt.scatter(Xtest,Ytest)
plt.plot(Xtest,Ypred)
```

### In this sales forecasting problem, both MSE and MAE can be used to evaluate model performance, but MAE is generally more suitable. This is because MAE measures the average absolute difference between predicted and actual sales in the same units (e.g., rupees or dollars), making it easier to interpret in a business context. Unlike MSE, which squares the errors and therefore gives disproportionately higher weight to large deviations, MAE treats all errors equally and is more robust to outliers, such as sudden unusual spikes or drops in monthly sales. Since in advertising and sales data such outliers are common, MAE provides a fairer and more practical measure of prediction accuracy, while MSE would be preferred only if large prediction errors are especially critical to avoid.
