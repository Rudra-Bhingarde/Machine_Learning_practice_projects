```python
import csv
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori , association_rules
```

```python
data =[]
with open("Market_Basket_Optimisation.csv") as file:
    reader= csv.reader(file,delimiter=',')
    for row in reader:
        data+=[row]
```

```python
data[:5]
```

```python
len(data)
```

```python
te = TransactionEncoder()
X = te.fit_transform(data)
```

```python
te.columns_
```

```python
df = pd.DataFrame(X,columns = te.columns_)
```

```python
df.head()
```

```python
frequent_items = apriori(df,min_support=0.02,use_colnames=True)
```

```python
frequent_items
```

```python
rules = association_rules(frequent_items,metric = 'confidence',min_threshold=0.1)
```

```python
rules
```

```python
rules = rules[['antecedents','consequents','support','confidence']]
```
